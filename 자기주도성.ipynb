```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
%matplotlib inline

df = pd.read_excel('자기주도적학습능력.xlsx', sheet_name = 0)
df.columns = ['before', 'after']
df.index = ['학생A', '학생B', '학생C', '학생D']
df.head()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='C:\Windows\Fonts\malgun.ttf').get_name()
rc('font', family=font_name)
df[['before', 'after']].plot(kind='box', title='계약학습 시행 전•후 자기주도적 학습능력')
```

```python
df['difference'] = df['after'] - df['before']
df['difference'].plot(kind='hist', title='difference')
```

```python
stats.probplot(df['difference'], plot= plt)
plt.title('difference Q-Q Plot')
plt.savefig('difference qq plot.png')
```

```python
stats.ttest_rel(df['before'], df['after'])
```

```python
import seaborn as sns
sns.distplot(df.iloc[:,0])
plt.show()
stats.probplot(df.iloc[:,0], plot=plt)
plt.show()
```
